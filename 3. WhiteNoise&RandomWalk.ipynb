#### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import acorr_ljungbox
```

#### Generating White Noise and Random Walk for comparision

```python
np.random.seed(0)
n = 1000

white_noise = np.random.normal(0,1,n)

random_shocks = np.random.normal(0,1,n)
random_walk = np.cumsum(random_shocks)
```

```python
plt.figure(figsize=(18,6))

plt.subplot(2,1,1)
plt.plot(white_noise, color ='#FF914D')
plt.title('White Noise')

plt.subplot(2,1,2)
plt.plot(random_walk, color ='#FF914D')
plt.title('Random Walk')

plt.show()
```

#### Ljung - Box Test

```python
lb_test_white_noise = acorr_ljungbox(white_noise, lags=[10], return_df=True)
print('Ljung-Box Test for White Noise: \n', lb_test_white_noise)
```

```python
lb_test_random_walk = acorr_ljungbox(random_walk, lags=[10], return_df=True)
print('Ljung-Box Test for Random Walk: \n', lb_test_random_walk)
```
